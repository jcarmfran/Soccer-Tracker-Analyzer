## Pre-Model Training

```python
model = YOLO("../models/yolov8m")

results = model.predict("../artifacts/input_videos/08fd33_4.mp4", save=True)
print(results[0]) # printing first frame
print("=================")
for box in results[0].boxes:
    print(box)
```

## Post-Model Training
Following training the model with YOLO version 8 (medium) over 100 epochs.

```python
# using the pre-trained model
model = YOLO("../models/best.pt")

results = model.predict("../artifacts/input_videos/08fd33_4.mp4", save=True)
print(results[0]) # printing first frame
print("=================")
for box in results[0].boxes:
    print(box)
```

```python
# using the pre-trained model
model = YOLO("../models/last.pt")

results = model.predict("../artifacts/input_videos/08fd33_4.mp4", save=True, project="../runs/detect", name="inference", exist_ok=True)
print(results[0]) # printing first frame
print("=================")
for box in results[0].boxes:
    print(box)
```
